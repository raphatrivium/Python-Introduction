1. Escreva um conjunto apropriadamente geral de funções que podem desenhar formas como na figura:

````python
import math
import ColabTurtle.Turtle
alice = ColabTurtle.Turtle
alice.initializeTurtle()
alice.speed(10)

#desenha um triangulo isosceles
def triangle(turtle, angle, height):
    turn_angle = 180 - (180 - angle)/2
    base_size = int(math.tan(math.radians(angle/2))*height)*2
    cathetus_size = int(height/math.cos(math.radians(angle/2)))
    turtle.forward(cathetus_size)
    turtle.left(turn_angle)
    turtle.forward(base_size)
    turtle.left(turn_angle)
    turtle.forward(cathetus_size)
    turtle.left(180)

def n_tagono(n, height):
    for i in range(n):
        triangle(alice, 360/n, height)

#Esquerda
alice.penup()
alice.left(90)
alice.forward(200)
alice.pendown()
n_tagono(5, 75)

#Meio
alice.penup()
alice.left(180)
alice.forward(200)
alice.left(90)
alice.pendown()
n_tagono(6, 75)

#Direita
alice.penup()
alice.right(90)
alice.forward(200)
alice.pendown()
n_tagono(7, 75)
````

1. Adicione instruções para desenhar o talo e folha da flor. Faça a refatoração apropriada para a interface da flor completa ficar "limpa".


````python
import math
import ColabTurtle.Turtle
alice = ColabTurtle.Turtle
alice.initializeTurtle()
alice.speed(10)

def polygon(turtle, length, n, circle_fraction):
    alfa = 360*circle_fraction/n
    for i in range(n):
        turtle.forward(length)
        turtle.left(alfa)

def circle(turtle, r, circle_fraction = 1):
    circumference = 2*math.pi*r
    n = 10
    length = int(circumference/n)
    polygon(turtle, length, n, circle_fraction)

def petal(turtle, circle_fraction):
    r = 10
    angle = 180 - circle_fraction*360
    circle(turtle, r, circle_fraction)
    turtle.left(angle)
    circle(turtle, r, circle_fraction)
    turtle.left(angle)

def flower(turtle, petals, circle_fraction):
    for i in range(petals):
        petal(turtle, circle_fraction)
        turtle.left(360/petals)

def stalk(turtle, height):
    turtle.forward(int(height*2/3))
    alice.left(100)
    petal(turtle, 1/5)
    alice.right(100)
    turtle.forward(int(height*1/3))

#Flor com caule
alice.left(180)
alice.color('green')
stalk(alice, 150)
alice.backward(150)
alice.left(180)

alice.color('red')
flower(alice, 7, 1/5)
````

````python
import math
import ColabTurtle.Turtle
alice = ColabTurtle.Turtle
alice.initializeTurtle()
alice.speed(10)

def polygon(turtle, length, n, circle_fraction):
    alfa = 360*circle_fraction/n
    for i in range(n):
        turtle.forward(length)
        turtle.left(alfa)

def circle(turtle, r, circle_fraction):
    circumference = 2*math.pi*r
    n = 5
    length = int(circumference/n)
    polygon(turtle, length, n, circle_fraction)

def spiral(turtle, size, circle_fraction):
    for i in range(size):
        circle(turtle, i, circle_fraction)

spiral(alice, 30, 1/8)
````
