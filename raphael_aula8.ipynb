1. Resolução numérica de equações diferenciais. Imagine que tem um/a maratonista treinando, que num trecho de $5 km$ tem velocidade quase constante de $12 km/h$. Produza uma lista distância percorrida em intervalos de 1 minuto. 

```
import matplotlib.pyplot as plt
def distance_OneMin(d,v,p):
  #input()
  t = 0;  S0 = 0;  S = 0
  dis_per = []
  tempo = []
  v = v/60 #km/min
  while S < d:
    S = S0 + v*t ;
    S = round(S, 3) #truncar valor para até 3 casas decimais
    dis_per.append(S) #adiciona item à lista
    tempo.append(t) #adiciona item à lista
    t = t+p
  print(dis_per)
  print(tempo)
  plt.plot(dis_per, tempo)
  plt.xlabel("distância m")
  plt.ylabel("tempo s ")
#distancia, velocidade, passo
distance_OneMin(5,12,1)
```

O/A mesmo/a atleta no seguinte trecho acelera  por $200$ metros até chegar em $15 km/h$  por $2 km$. Queremos obter uma tabela ou gráfico da distância percorrida en função do tempo e o tempo total para chegar em $7 km$, faça isto utilizando listas. 

```
import matplotlib.pyplot as plt
# Calcular as Raízes de uma Equação do 2º Grau
def raizes(a, b, c):
    D = (b**2 - 4*a*c)
    x1 = (-b + D**(1/2)) / (2*a)
    x2 = (-b - D**(1/2)) / (2*a)
    print('\nValor da raiz 1: {0}'.format(x1))
    print('Valor da raiz 2: {0}'.format(x2))
    return x1

def distance_OneMin(d,deltaS,d2,v0,vf,p):
  t = 0;  S0 = 0;  S = 0
  dis_per = []
  tempo = []
  v0 = v0/60 #km/min
  vf = vf/60 #km/min
  while S < d:
    S = S0 + v0*t
    tempo.append(t) #adiciona item à lista
    S = round(S,3) #truncar valor para até 3 casas decimais
    dis_per.append(S) #adiciona item à lista
    t = t+p #passo

  t1 = (S-S0)/v0
  a = ( (vf**2) -(v0**2))/(2*deltaS) #Equação de Torricelli   
  S1 = S + v0*(t-t1) +(1/2)*(a)*((t-t1)**2)
  while (S1-S) < deltaS:     
    S1 = S + v0*(t-t1) +(1/2)*(a)*((t-t1)**2)
    dis_per.append(round(S1,3))
    tempo.append(t)
    t = t+p

  t2 = raizes( round(a/2,3), v0, -deltaS )
  S2 = S+deltaS
  while S2 < d2:
    S2 = S+deltaS + vf*(t-t1+t2)
    tempo.append(t)
    dis_per.append(round(S2,3))
    t = t+p   
  print(dis_per)
  print(tempo)
  plt.plot(dis_per, tempo)
  plt.xlabel("distância m")
  plt.ylabel("tempo s ")
#d1,deltaS,d2,v0,vf,p
distance_OneMin(5, 0.2, 7, 12, 15, 1)

####!!!!! Termos acima de 7 não deveriam aparecer
```
